### HS코드를 BEC로 매핑하기

```python
Korea_export = pd.read_csv(directory,encoding='unicode_escape')
Korea_export['Code'] = Korea_export['Code'].apply(lambda x: str(x) if len(x)==6 else '0'+str(x))
Korea_export = Korea_export.drop(0, axis=0)
```

```python
Inner_data = pd.merge(Korea_export, HS_to_BEC, left_on='Code', right_on='HS6', how='inner')
```

### 수입액 기준 전체 나라 총합

```python
HS_to_BEC = pd.read_excel('HS2012-17-BEC5 -- 08 Nov 2018.xlsx', sheet_name=1)
HS_to_BEC['HS6'] = HS_to_BEC['HS6'].astype(str)
HS_to_BEC['HS6'] = HS_to_BEC['HS6'].apply(lambda x: str(x) if len(x)==6 else '0'+str(x))
```

```python
data_dir = os.listdir()
```

```python
data_dir = data_dir[11]
```

```python
for case_name in os.listdir(data_dir):
    current_path = os.path.join(data_dir, case_name)
    count = 0
    if current_path.split('_')[6]=='imported':
        if count == 0:
            print(current_path)
            temp = pd.read_csv(current_path, encoding='unicode_escape')
            new_df = temp.drop(0, axis=0)
            new_df['Code'] = new_df['Code'].apply(lambda x: str(x) if len(x)==6 else '0'+str(x))
            new_df = pd.merge(new_df, HS_to_BEC, left_on='Code', right_on='HS6', how='inner')
            summury=new_df
            count+=1
        else:
            print(current_path)
            temp = pd.read_csv(current_path, encoding='unicode_escape')
            new_df = temp.drop(0, axis=0)
            new_df['Code'] = new_df['Code'].apply(lambda x: str(x) if len(x)==6 else '0'+str(x))
            new_df = pd.merge(new_df, HS_to_BEC, left_on='Code', right_on='HS6', how='inner')
            summury=summury.add(new_df, fill_value = 0, axis = 0)
```

```python
summury = summury.groupby('BEC5Code1').sum()
summury
```

```python
columns_list = summury.columns[0:10]
```

```python
for i in range(10):
    a=2012
    a=(a+i)
    ratio = summury['Imported value in '+str(a)]
    labels = summury.index
    #plt.pie(ratio, labels=labels, autopct='%.1f%%')
    plt.title('Imported value in '+str(a))
    wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
    plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, wedgeprops=wedgeprops)
    plt.figure(figsize=(20,20))
    plt.show()
```

```python
summury_im = summury[columns_list]
summury_im
```

```python
index_large = summury_im.T.sum().sort_values(ascending=False).index[0:10]
```

```python
set_image = summury_im.loc[index_large]
```

```python
set_image
```

```python
set_image.div(set_image.sum(axis=0), axis=1).T.plot(figsize=(16, 8))
```
